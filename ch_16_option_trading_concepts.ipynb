```python
import random

import numpy as np
import seaborn as sns
```

# Path Dependence  
pg 266-7

```python
start_price = 100
seq = [0.01, 0.05, -0.01, -0.05]
```

```python
sims = 10
paths = {}
for i in range(0, sims):
    random.shuffle(seq)
    p = start_price
    paths[i] = [p]
    for ret in seq:
        p = p * (1 + ret)
        paths[i].append(p)
```

```python
sns.lineplot(data=paths)
```

## 252 day example

```python
seq1 = np.random.normal(0, 0.01, 126)
seq2 = np.negative(seq1)
seq = np.concatenate((seq1, seq2))
```

```python
sims = 106
paths = {}
for i in range(0, sims):
    random.shuffle(seq)
    p = start_price
    paths[i] = [p]
    for ret in seq:
        p = p * (1 + ret)
        paths[i].append(p)
```

```python
subset = dict((k, paths[k]) for k in range(0, 8))
```

```python
sns.lineplot(data=subset)
```

```python
print("Final price: ", round(paths[0][-1], 2))
```

```python
max_val = 0
min_val = 100
for path in paths:
    mx = max(paths[path])
    mn = min(paths[path])
    if mx > max_val:
        max_val = mx
    if mn < min_val:
        min_val = mn
```

```python
print("High from all paths: ", max_val)
print("Low from all paths: ", min_val)
```
