```python
import math
import pandas as pd
```

### Calculating Historical Vol

```python
data = {'Day': range(1, 23), 
        'Price': [99.36, 98.73, 98.97, 99.94, 100.20, 101.47, 102.06, 101.5, 102.45, 102.89,
                  103.09, 103.71, 104.04, 105.17, 104.52, 104.32, 103.31, 102.27, 101.63, 101.22,
                  101.44, 101.04]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
nl_ret = [math.nan]
sq_nl = [math.nan]
prev = None
for index, row in df.iterrows():
    if index == 0:
        prev = row['Price']
        continue
    nl_ret.append(math.log(row['Price']/prev))
    sq_nl.append(math.log(row['Price']/prev)**2)
    prev = row['Price']
```

```python
df['Log'] = nl_ret
df['Log_square'] = sq_nl
```

```python
df
```

```python
annualized_vol = math.sqrt(df['Log_square'].iloc[1:].sum()/(len(df)-1))*math.sqrt(248)
print(annualized_vol*100)
```

### Filtering

```python
decay_factor = .97
```

```python
decays = []

for i in range(1, len(df)):
    decays.append(decay_factor**i)
decays.append(math.nan)
```

```python
decays.reverse()
df['Decay'] = decays
```

```python
df
```

```python
weighted_square_rets = [math.nan]
for index, row in df.iterrows():
    if index == 0:
        continue
    weighted_square_rets.append(row['Decay'] * row['Log_square'])
```

```python
df['weighted_square_rets'] = weighted_square_rets
```

```python
df
```
